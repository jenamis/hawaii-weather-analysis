```python
# Import dependencies
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Import Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# Set up connection to SQLite database
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# Reflect an existing database into a new model
Base = automap_base()

# Reflect tables
Base.prepare(engine, reflect = True)
```

```python
# View all classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create session (link) from Python to DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design query to retrieve last 12 months of precipitation data and plot results
# Start from last data point in database, calculate date one year from last date in data set
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days = 365)
```

```python
# Perform a query to retrieve the data and precipitation scores
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()
print(results)
```

```python
# Save query results as a Pandas DF and set index to date column
df = pd.DataFrame(results, columns = ["date", "precipitation"])
df.set_index(df["date"], inplace = True)

# Print DF with index
# print(df)

# Print DF without index
print(df.to_string(index = False))
```

```python
# Sort DF by date
df = df.sort_index()
print(df.to_string(index = False))
```

```python
# Use Pandas Plotting with Matplotlib to plot data
df.plot(xlabel = "Date", ylabel = "Total Precipitation", title = "Total Precipitation per Day", figsize = (20, 10))
plt.show()
```

```python
# Use Pandas to calcualte summary statistics for precipitation data
df.describe()
```

```python
# Determine number of stations in dataset
session.query(func.count(Station.station)).all()
```

```python
# Determine most active stations
# List stations and measurement counts in descending order
session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()
```

```python
# Calculate the lowest, highest, and average temperature for most active station
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
    filter(Measurement.station == "USC00519281").all()
```

```python
# Query last 12 months of temperature data for station with highest number of temperature observations
results = session.query(Measurement.tobs).\
        filter(Measurement.station == "USC00519281").\
        filter(Measurement.date >= prev_year).all()
print(results)
```

```python
# Convert temperature results to DF
df = pd.DataFrame(results, columns = ["tobs"])
print(df.to_string(index = False))
```

```python
# Plot temperature results as histogram
df.plot.hist(bins = 12, title = "Distribution of Temperature Observations for Station USC00519281", figsize = (20, 10))
plt.tight_layout()
plt.show()
```
